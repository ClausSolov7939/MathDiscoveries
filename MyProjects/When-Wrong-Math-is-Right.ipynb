## Произведение матрицы

### Введение

Случайно наткнувшись на этот мем, где найденое произведение матриц выполнено *неправильно* ([произведение Адамара](https://en.wikipedia.org/wiki/Hadamard_product_(matrices))), но ответ получился правильный

$$
\begin{bmatrix} 3 & 6 \\ 0 & 1 \end{bmatrix} 
\begin{bmatrix} 5 & 4 \\ 0 & 2 \end{bmatrix} = 
\begin{bmatrix} 15 & 24 \\ 0 & 2 \end{bmatrix} 
$$

И возник вопрос: **Когда это правдиво?**

### Исследование

Допустим такое произведение матрицы и напишем равенство

$$
\begin{bmatrix} a & b \\ c & d \end{bmatrix} 
\begin{bmatrix} e & f \\ g & h \end{bmatrix} = 
\begin{bmatrix} ae & bf \\ cg & dh \end{bmatrix} 
$$

Тогда для него распишем систему уравнений

$$
\begin{cases}
\cancel{ae} = \cancel{ae} + bg \\
bf = af + bh \\
cg = ce + dg \\
\cancel{dh} = cf + \cancel{dh}
\end{cases}
$$

$$
\begin{cases}
bg = 0 \\
cf = 0 \\
bf = af + bh \\
cg = ce + dg
\end{cases}
$$

Из первых двух строк выводятся 4 условия:
1. $b = 0$ и $c = 0$
4. $g = 0$ и $f = 0$
2. $b = 0$ и $f = 0$
3. $g = 0$ и $c = 0$

Рассмотрим последние две строки при каждом условии:
1. При $b = 0$ и $c = 0$

$$
\begin{cases}
af = 0\\
dg = 0
\end{cases}
$$

2. При $g = 0$ и $f = 0$

$$
\begin{cases}
bh = 0 \\
ce = 0 
\end{cases}
$$

При условиях 1 и 2 одна из матриц имеет хотя бы 3 нуля, что не сильно интересует

3. При $b = 0$ и $f = 0$

$$
cg = ce + dg
$$

$$
\dfrac{e}{g} + \dfrac{d}{c} = 1
$$

4. При $g = 0$ и $c = 0$

$$
bf = af + bh
$$


$$
\dfrac{a}{b} + \dfrac{h}{f} = 1
$$

При условиях 3 и 4 обе матрицы имеют один нуль на одинаковых позициях и имеет зависимость элементов. 

Другими словами

$$
\begin{bmatrix} a & b \\ 0 & d \end{bmatrix} 
\begin{bmatrix} e & f \\ 0 & h \end{bmatrix} = 
\begin{bmatrix} ae & bf \\ 0 & dh \end{bmatrix}, \;

\dfrac{a}{b} + \dfrac{h}{f} = 1
$$

$$
\begin{bmatrix} a & 0 \\ c & d \end{bmatrix} 
\begin{bmatrix} e & 0 \\ g & h \end{bmatrix} = 
\begin{bmatrix} ae & 0 \\ cg & dh \end{bmatrix}, \;

\dfrac{e}{g} + \dfrac{d}{c} = 1
$$

## Сложение дробей

При каких условия сложение дробей становится [Медиантой](https://en.wikipedia.org/wiki/Mediant_(mathematics)).

$$
\dfrac{a}{b}+\dfrac{c}{d} = \dfrac{a+c}{b+d}
$$

$$
\dfrac{a+c}{b+d} = \dfrac{ad+cb}{bd}
$$

$$
\dfrac{a+c}{b+d} = \dfrac{ad+cb}{bd}
$$
