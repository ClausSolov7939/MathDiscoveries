## Эквивалентности
$$
\begin{array}{l}
P_{n}\cdot S^{-1}_{n}=x_{2}x_{3}\dots x_n+x_{1}x_{3}\dots x_{n}+x_1x_{2}x_{4}\dots x_n\\
S_{n}^{-1}=\dfrac{1}{x_{1}}+\dfrac{1}{x_{2}}+\dots+\dfrac{1}{x_{n}}\\
S_{n}^{1}\cdot S_{n}^{-1}=\sum\limits^{n}_{k=1}{\dfrac{S_{n}^{1}}{x_{k}}}\text{ <- логично =)}
\end{array}
$$

# История появления

Во третьем семестре появляется **Вычислительная математика**, в которой был полином Лагранжа, который представлял собой полином с коэффициентами в виде дробей произведений скобок уравнения $n$-ого порядка, которую нужно было преобразовать.
$$
\dfrac{(x-x_1)(x-x_2)(x-x_{3})(x-x_{4})}{(x-0)(x-0.5)(x-1)(x-1.5)}
$$
Самым "простым" способом преобразовать скобки в полином было *перемножение скобок*, что оказалось достаточно **муторным** занятием.

Но после обобщенного изучения выяснилось, что **коэффициенты очень просто выражать**.

$$
\begin{array}{r}
(x-x_{1})(x-x_{2})(x-x_{3})(x-x_4)=\\
1\cdot x^{4}\\
-(x_{1}+x_{2}+x_{3}+x_{4})\cdot x^{3}\\
+(x_{1}x_{2}+x_{1}x_{3}+x_{1}x_{4}+x_{2}x_{3}+x_{2}x_{4}+x_{3}x_{4})\cdot x^{2}\\
-(x_{1}x_{2}x_{3}+x_{1}x_{2}x_{4}+x_{1}x_{3}x_{4}+x_{2}x_{3}x_{4})\cdot x^{1}\\
+(x_{1}x_{2}x_{3}x_{4})\cdot x^{0}
\end{array}
$$
Дальше все рассуждения подтянули обобщенные формулы суммы и произведения членов.
