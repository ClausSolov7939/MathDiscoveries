# Информаторий

## Описание

Полуавтоматический интернет-сервис по заполнению истории, достижений и заслуг человека. Фактически, Wikipedia для каждого человека. Также можно отобразить генеалогические деревья.

## Мысли
- Записывать созданные проекты и моменты. События связанные с людьми нужно прилагать с ссылками на них.


- Запиши свою страницу истории
	1. Записывай события с датой и местоположением и по возможности с отзывом.
	2. Автоматическая запись местоположения и времени.
- Создать **всемирную базу данных о людях**, включающее их **здоровье**, **контактную информацию**, **достижений**, **планов на ближайшее будущее**, местоположение. "Никто не забыт, никто не потерян”
- Единый личный аккаунт человека:
	- **Календарь планов**, вместе с адресами мест мероприятий. Синхронизация с планированием другого человека или группы.
	- **Карта для расчета пути и времени**, связан с календарем для расчёта подходящего времени для планирования. Карта всех видов транспортов с будильником для остановок общ. транспорта. В общ. транспорте должны знать до какой остановки кто едет, чтобы уступать и занимать места осознанно.
	- **Средства связи**: чаты (почты), звонки, видеозвонки
	- **Социальные сети** для отзывов и мнений что либо изменить к лучшему. Отзывы к людям, системе управления, общ. транспорта. Также в соц. сети представлены **статусы** людей: здоровье, настроение. Все должны точно знать, кто чем болеет или болел по информации врачебной карты.
	- Другие средства массовой информации. 
	- Какое либо устройство создаётся, всегда перед использованием происходит полноценная синхронизация данных. 
	- "Твой телефон общий. Каждый может им пользоваться. Но будет использовать только свои данные во имя защиты авторских прав человека"
	- Никто не скроет своих преступлений, все о них будут известны.
	- Разве что только количество денег не будет общедоступно, но уверен, что вскоре мы от них откажемся.	    
	- С рождения человека данные о нем добавляются в всемирную базу данных людей и создают аккаунт в всемирной соц. сети.
		1. Дети  первых лет носят браслеты с постоянной геолокацией на ноге и руке. Никаких взаимодействий с устройство ребенок не имеет.
	        2. Дети дошкольного возраста носят усовершенные браслет, которые ещё измеряют температуру, сердцебиение, кислород в крови, давление.
	        3. Дети школьного возраста носят браслеты с дисплеем
	- Всякое ежедневное потребление биологический ресурсов, в последствие поглощение его организмом, должно быть расчитано и быть выдано ежедневно. Так как организму не нужно  слишком много или слишком мало, а нужно в расчитанном кол-ве, для поддержания здоровья организма.
	   
	- Устройство, которое работает только после предоставления биометрических данных, имеет геолокацию и связь других устройств. Устройство ежечасно
	    
	    питание, поэтому работоспособность устройство зависит от здоровья владельца. Можно оповестить пользователя и проверить жив ли он.
	    
	- Цель любого человека:
	    
	    1. Выжить и прожить долгую жизнь
	    2. Оставить после себя потомство
	- Причины смерти:
	    
	    1. Природа (болезнь, катаклизм), от которой постоянно спасаются 
	    2. Самоубийство (из-за физических или химических реакций). Фактический **проигрыш** в игре Жизнь.
	    3. Нежеланное убийство от кого-то, кто абсолютно не хотел этого.	    
	    4. Намеренное убийство от кого-то, кто это желал и стремился к этой цели.
	    5. (!) Наёмное убийство от кого-то, кто не хотел, но ему пришлось из-за второго.
	    
	- Первый совершил нежеланное убийство, второй совершил намеренное убийство.
	    - Степени по возрастанию:
		    1. Пропал без вести (неизвестно)
		    2. Природа
		    3. Нежеланное
		    4. Самоубийство
		    5. Намеренное
    

# Методика философии

Желательно продумать теорию перед практикой для предотвращения **плохих последствий** при и после реализации.
