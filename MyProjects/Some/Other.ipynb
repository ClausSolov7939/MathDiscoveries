# Общие суждения
- Главная цель обучения в ВУЗе:
	- Отсрочка
	- Книга о высшем образовании
	- Стипендия
- Обучаться нужно легко, чтобы время оставалось на самообучения
- Автоматы выгодны преподавателям

# План
Задать вопрос лекторам и преподавателям. 
- Есть ли автомат:
	- Если **есть**, то что делать для **него**
	- Если **нет**, то что делать для **допуска**
- Выяснить что делать для допуска/автомата.
	- **P** - рукописный документ \[Конспект, лабораторная работа, типовой расчет\]
	- **E** - электронный документ \[Документ, программа, видеоматериал\]
	- **O** - устная работа \[Топики, выступления\]
	- **S** - умения (уметь решить контрольную работы)
	- **A** - присутствие на парах 
- Какие сроки сдачи работ:
	- **JIT** (just in time) - на месте \[контрольные работы\]
	- **UTD** (until the deadline) - до определенного срока \[практики\]
- Выяснить как будет проводится экзамен/зачет:
	- **P** - письменно (решение задач, тест)
	- **O** - устный ответ
	- **E** - электронный тест

1. Проект "Каталог программ"
	- Прописанная точная программа обучения
	- Программа не закреплена за преподавателем
	- Есть рейтинг преподавателей
	- Есть контроль версий методичек и комментария к ним
2. Проект "Гибридные лекции"
	- 80% - заочные лекции, 20% - очные
	- Плюсы:
		- Больше 1 потока за пару
		- Живые выступления не отменены
