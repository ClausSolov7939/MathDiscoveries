## Замечательные пределы
1. $\lim\limits_{x\to0}{\dfrac{\sin(x)}{x}}=1$
2. $\lim\limits_{x\to\infty}{\left(1+\dfrac{1}{x}\right)^{x}}=\lim\limits_{x\to0}(1+x)^\dfrac{1}{x}=e$
## Методы нахождения пределов
1. 
$$
\begin{array}{l}
\lim\limits_{x\to\infty}{\dfrac{P_{3}(x)}{P_{2}(x)}}=
\lim\limits_{x\to\infty}{\dfrac{\dfrac{P_{3}(x)}{x^{2}}}{\dfrac{P_{2}(x)}{x^{2}}}}=
\lim\limits_{x\to\infty}{\dfrac{a\cdot x}{b}}=\infty\\
\lim\limits_{x\to\infty}{\dfrac{P_{2}(x)}{P_{3}(x)}}=
\lim\limits_{x\to\infty}{\dfrac{\dfrac{P_{2}(x)}{x^{2}}}{\dfrac{P_{3}(x)}{x^{2}}}}=
\lim\limits_{x\to\infty}{\dfrac{a}{b\cdot x}}=0\\
\end{array}
$$
2. 
$$
\lim\limits_{x\to\infty}{\sqrt{f(x)}-\sqrt{g(x)}}=
\lim\limits_{x\to\infty}{\dfrac{f^{2}(x)+g^{2}(x)}{\sqrt{f(x)}+\sqrt{g(x)}}}
$$
3. Правило Лопиталя $\left[\dfrac{f(x)}{g(x)}=\dfrac{\infty}{\infty},\dfrac{0}{0}\right]$ $$\lim\limits_{x\to\infty}{\dfrac{f(x)}{g(x)}}=\lim\limits_{x\to\infty}{\dfrac{f'(x)}{g'(x)}}$$

## Ряд Лорана

$$
\sum\limits_{n=-\infty}^{\infty}{c_{n}(z-z_{0})^{n}}=
\sum\limits_{n=1}^{\infty}{\dfrac{c_{-n}}{(z-z_{0})^{n}}}+
\sum\limits_{n=0}^{\infty}{c_{n}(z-z_{0})^{n}}
$$
Правильная часть (Ряд Тейлора): $\sum\limits_{n=0}^{\infty}{c_{n}(z-z_{0})^{n}}$
