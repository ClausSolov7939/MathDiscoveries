# Выражение суммы степеней

https://en.wikipedia.org/wiki/Sums_of_powers
$$
\sum\limits_{k=1}^{n}{k^{1}}=\dfrac{n(n+1)}{2}
$$
$$
\begin{array}{rcl}
(n+1)^{3}&=& n^{3}+3\cdot n^{2}+3\cdot n+1\\
2^{3}&=& 1^{3}+3\cdot 1^2+3\cdot1+1\\
3^{3}&=& 2^{3}+3\cdot 2^2+3\cdot2+1\\
\vdots & & \vdots\\
n^{3}&=& (n-1)^{3}+3(n-1)^{2}+3(n-1)+1
\end{array}
$$
Тогда
$$\small
\begin{array}{l}
\cancel{2^{3}+3^{3}+\dots+n^{3}}+(n+1)^{3}=\\(1^{3}+\cancel{2^{3}+\dots+(n-1)^{3}+n^{3}})+3\boxed{\sum\limits_{k=1}^{n}{k^{2}}}+3\sum\limits_{k=1}^{n}{k}+n\cdot1
\end{array}
$$
$$
\begin{array}{l}
(n+1)^{3}=1+3\boxed{\sum\limits_{k=1}^{n}{k^{2}}}+3\sum\limits_{k=1}^{n}{k}+n\cdot1\\
\sum\limits_{k=1}^{n}{k^{2}}=\dfrac{(n+1)^{3}-(n+1)-3\sum\limits_{k=1}^{n}{k}}{3}\\
\sum\limits_{k=1}^{n}{k^{2}}=\dfrac{(n+1)^{3}-(n+1)-3\dfrac{n\cdot(n+1)}{2}}{3}\\
\sum\limits_{k=1}^{n}{k^{2}}=(n+1)\dfrac{(n+1)^{2}-1-3\dfrac{n}{2}}{3}\\
\sum\limits_{k=1}^{n}{k^{2}}=(n+1)\dfrac{2n^{2}+4n-2+2-3n}{6}\\
\boxed{\sum\limits_{k=1}^{n}{k^{2}}=\dfrac{n(n+1)(2n+1)}{6}=\dfrac{n(n+1)}{3}\cdot\left(n+\dfrac{1}{2}\right)}
\end{array}
$$
Таким же методом можно выразить и остальные формулы частных сумм.

$$
\begin{array}{l}
\sum\limits_{k=1}^{n}{k^{1}}=\dfrac{n(n+1)}{2}\\
\sum\limits_{k=1}^{n}{k^{2}}=\dfrac{n(n+1)}{3}\cdot\left(n+\dfrac{1}{2}\right)\\
\sum\limits_{k=1}^{n}{k^{3}}=\dfrac{n(n+1)}{4}\cdot\left(n^{2}+n\right)\\
\sum\limits_{k=1}^{n}{k^{4}}=\dfrac{n(n+1)}{5}\cdot\left(n^{3}+\dfrac{3}{2}n^{2}+\dfrac{1}{6}n-\dfrac{1}{6}\right)\\
\sum\limits_{k=1}^{n}{k^{5}}=\dfrac{n(n+1)}{6}\cdot\left(n^{4}+2n^{3}+\dfrac{1}{2}n^2-\dfrac{1}{2}n\right)\\
\sum\limits_{k=1}^{n}{k^{6}}=\dfrac{n(n+1)}{7}\cdot\left(n^{5}+\dfrac{5}{2}n^{4}+n^{3}-n^2-\dfrac{1}{6}n+\dfrac{1}{6}\right)\\
\sum\limits_{k=1}^{n}{k^{7}}=\\
\sum\limits_{k=1}^{n}{k^{8}}=\\
\sum\limits_{k=1}^{n}{k^{9}}=\\
\sum\limits_{k=1}^{n}{k^{10}}=\\
\end{array}
$$

# Обобщенная формула

$$\LARGE
(n+1)^{\alpha}-(n+1)=\sum\limits_{m=1}^{\alpha-1}{\left(\binom{a}{m}\cdot\boxed{\sum\limits_{k=1}^{n}{k^{\alpha-m}}}\right)}
$$
$$\LARGE
\begin{array}{l}
\sum\limits_{k=1}^{n}{k^{\alpha-1}}=\dfrac{(n+1)^\alpha-(n+1)-\sum\limits_{m=2}^{\alpha-1}{\left[\dbinom{\alpha}{m}\sum\limits_{k=1}^{n}{(k^{\alpha-m})}\right]}}{\alpha}=\\\small

\dfrac{(n+1)^{\alpha}}{\alpha}-
\dfrac{(n+1)}{\alpha}-
\dfrac{\dbinom{\alpha}{2}\sum\limits_{k=1}^{n}(k^{\alpha-2})}{\alpha}-
\dfrac{\dbinom{\alpha}{3}\sum\limits_{k=1}^{n}(k^{\alpha-3})}{\alpha}-
\dots-
\dfrac{\dbinom{\alpha}{\alpha-2}\sum\limits_{k=1}^{n}(k^{2})}{\alpha}-
\dfrac{\dbinom{\alpha}{\alpha-1}\sum\limits_{k=1}^{n}(k^{1})}{\alpha}=\\
\small
\dfrac{(n+1)^{\alpha}}{\alpha}-
\dfrac{(n+1)}{\alpha}-
\dfrac{\dbinom{\alpha}{\alpha-1}\sum\limits_{k=1}^{n}(k^{1})}{\alpha}-
\dfrac{\dbinom{\alpha}{\alpha-2}\sum\limits_{k=1}^{n}(k^{2})}{\alpha}-
\dots-
\dfrac{\dbinom{\alpha}{3}\sum\limits_{k=1}^{n}(k^{\alpha-3})}{\alpha}-
\dfrac{\dbinom{\alpha}{2}\sum\limits_{k=1}^{n}(k^{\alpha-2})}{\alpha}=\\
\small
\dfrac{(n+1)^{\alpha}}{\alpha}-
\dfrac{(n+1)}{\alpha}-
\dfrac{\alpha\sum\limits_{k=1}^{n}(k^{1})}{a}-

\dfrac{(n+1)^{\alpha}}{\alpha}-

\dfrac{\prod\limits_{j=1}^{2}{(a-j)}}{3!}\sum\limits_{k=1}^{n}(k^{3})
\dots=
\end{array}
$$
