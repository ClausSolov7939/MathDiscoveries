> Если площадь круга $\pi\cdot r^{3}$, то площадь какого правильного многоугольника равняется $3\cdot r^{3}$

# Площадь любого правильного многоугольника
Найдем общую формулу нахождения площадь правильного многоугольника через его радиус и кол-во сторон.

Площадь треугольника по двум равным сторонам $r$ и углу $\alpha$ между ними равен:
$$S_{\Delta}=\dfrac{r^{2}\sin{\alpha}}{2}$$
В любой правильный многоугольник со сторонами в кол-ве $n$, можно построить с помощью $n$ равнобедренных треугольников.

Тогда общая формула будет:

$$
S=r^{2}\cdot\dfrac{n\cdot\sin\left(\frac{2\pi}{n}\right)}{2}
$$
Из этого следует, что при бесконечном большом кол-ве сторон, многоугольника становится **кругом**.

$$
\lim_{n\to\infty}{\dfrac{n\cdot\sin{\left(\frac{2\pi}{n}\right)}}{2}}=
\lim_{n\to\infty}{\dfrac{\cancel{n}\cdot{\left(\frac{2\pi}{\cancel{n}}\right)}}{2}}=\pi
$$
---
Теперь, у какого многоугольника площадь равна $3r^{2}$?
**Ответ**: у **$12$-угольника**
$$\dfrac{12\cdot\sin\left(\frac{2\pi}{12}\right)}{2}=6\cdot\sin\left(\frac{\pi}6\right)=6\cdot\frac12=3$$
Это необычно, как разница от $3$ до $\pi=3.1415...$ выражается от $12$-угольника до круга ($\infty$-угольника).

---
А у какого многоугольника площадь равна $2r^{2}$?
**Ответ**: у $4$-угольника (квадрата)
$$
\dfrac{4\cdot\sin\left(\frac{2\pi}{4}\right)}{2}=2\cdot\sin\left(\frac{\pi}2\right)=2\cdot1=2
$$
---
