# Главные определения

$$\LARGE
i=\sqrt{-1}
$$
$$\LARGE
\begin{array}{l}
\begin{array}{l}
z=x+iy\to&\vec{a}=\begin{pmatrix} x\\y \end{pmatrix}\\
i\cdot z=-y+ix\to&\vec{b}=\begin{pmatrix}-y\\x\end{pmatrix}\\
i^2\cdot z=-z=-x-iy\to&\vec{c}=\begin{pmatrix}-x\\-y\end{pmatrix}\\
i^3\cdot z=-iz=y-ix\to&\vec{d}=\begin{pmatrix}y\\-x\end{pmatrix}\\
\end{array}\\
\vec{a}\cdot\vec{b}=0\to\vec{a}\bot\vec{b}\\
\vec{b}\cdot\vec{c}=0\to\vec{b}\bot\vec{c}\\
\vec{c}\cdot\vec{d}=0\to\vec{c}\bot\vec{d}\\
\vec{a}\bot\vec{b}\bot\vec{c}\bot\vec{d}
\end{array}
$$
Домножение на $i$ поворачивает вектор числа на $90\degree$

# Уравнение Эйлера
$$\LARGE
e^{ix}=\cos{(x)}+i\cdot\sin(x)
$$
## Доказательство по Рядам Тейлора

>Доказывает почему в уравнении стоит $e$
$$\small
\begin{array}{l}
e^{x}=\sum\limits_{n=0}^{\infty}{\dfrac{x^{n}}{n!}}=1+x+\dfrac{x^{2}}{2!}+\dfrac{x^{3}}{3!}+\dfrac{x^{4}}{4!}+\dfrac{x^{5}}{5!}+\dots\\
\boxed{e^{ix}=\sum\limits_{n=0}^{\infty}{\dfrac{(i\cdot x)^{n}}{n!}}=1+ix-\dfrac{x^{2}}{2!}-\dfrac{i\cdot x^{3}}{3!}+\dfrac{x^{4}}{4!}+\dfrac{i\cdot x^{5}}{5!}-\dfrac{x^{6}}{6!}+\dots}\\
\boxed{\cos(x)=\sum\limits_{n=0}^{\infty}{(-1)^{n}\dfrac{x^{2n}}{(2n)!}}=1-\dfrac{x^{2}}{2!}+\dfrac{x^{4}}{4!}-\dfrac{x^{6}}{6!}+\dfrac{x^{8}}{8!}+\dots}\\
\sin(x)=\sum\limits_{n=0}^{\infty}{(-1)^{n}\dfrac{x^{2n+1}}{(2n+1)!}}=x-\dfrac{x^{3}}{3!}+\dfrac{x^{5}}{5!}-\dfrac{x^{7}}{7!}+\dfrac{x^{9}}{9!}+\dots\\
\boxed{i\cdot\sin(x)=i\cdot x-\dfrac{i\cdot x^{3}}{3!}+\dfrac{i\cdot x^{5}}{5!}-\dfrac{i\cdot x^{7}}{7!}+\dfrac{i\cdot x^{9}}{9!}+\dots}
\end{array}
$$


## Принятие и способы преобразования

> Доказывает почему в уравнении стоят $\cos(x)$ и $\sin(x)$

$$
\begin{array}{l}
e^{ix}=\cos(x)+i\cdot\sin(x)\\
e^{-ix}=\cos(x)-i\cdot\sin(x)
\end{array}
$$
$$
\begin{array}{l}
\cos(x)=\dfrac{e^{ix}+e^{-ix}}{2}&
\sin(x)=\dfrac{e^{ix}-e^{-ix}}{2i}
\end{array}
$$
Это, блин, работает =D


# Функции комплексной переменной

- $e^{z}=e^{x+iy}=e^{x}\cdot(\cos(y)+i\cdot\sin(y))$
- $\text{Ln}(z)=\ln(|z|)+i\cdot\text{Arg}(z)=\ln|z|+i\cdot(\arg(z)+2\pi k), k\in \mathbb{Z}$
- $\cos(z)=\dfrac{e^{iz}+e^{-iz}}{2}$
- $\sin(z)=\dfrac{e^{iz}-e^{-iz}}{2i}$
